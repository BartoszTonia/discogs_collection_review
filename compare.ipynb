```python
import matplotlib.pyplot as plt
import pandas as pd

dfA = pd.read_csv('out/collection_280421.csv')
dfB = pd.read_csv('out/collection_030621.csv')

dfA['id'].count(), dfB['id'].count()
```

```python
df = pd.merge(dfA, dfB, how='inner', on=['id', 'title', 'url'])
df = df[(df['last_sold_x'] != 0) & (df['last_sold_y'] != 0)]
```

```python
view = ['id','title', 'med_x', 'med_y', 'url', 'for_sale_x', 'for_sale_y']
df.query('start_x==0 | start_y==0')[view]
```

```python
df['med_diff'] = df['med_y'] - df['med_x']
df['med_diff'].describe()
```

```python
view = ['id','title', 'med_x', 'med_y', 'url', 'for_sale_x', 'for_sale_y']
df.sort_values(by='med_diff', ascending=True)[view].head(10)
```

```python
df['high_diff'] = df['high_y'] - df['high_x']
df['high_diff'].describe()
```

```python
view = ['id','title', 'high_x', 'high_y', 'url', 'for_sale_x', 'for_sale_y']
df.sort_values(by='high_diff', ascending=True)[view].head(10)
```

```python
df['low_diff'] = df['low_y'] - df['low_x']
```

```python
view = ['id','title', 'low_x', 'low_y', 'url', 'for_sale_x', 'for_sale_y']
df.sort_values(by='low_diff', ascending=False)[view].head(10)
```

```python
view = ['id','title', 'low_x', 'med_x', 'high_x', 'low_y', 'med_y', 'high_y', 'url', 'stat_sum']
df['stat_sum'] = df['low_diff'] + df['med_diff'] + df['high_diff']
df.sort_values(by='stat_sum', ascending=False)[view]
```
